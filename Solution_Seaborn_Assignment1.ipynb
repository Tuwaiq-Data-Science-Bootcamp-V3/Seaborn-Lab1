## Data Description

- Survival : 0 = No, 1 = Yes
- pclass(Ticket Class) : 1 = 1st, 2 = 2nd, 3 = 3rd
- Sex(Gender) : Male, Female
- Age : Age in years
- SibSp : Number of siblings/spouses abroad the titanic
- Parch : Number of parents/children abrod the titanic
- Ticket : Ticket Number
- Fare : Passenger fare
- Cabin : Cabin Number
- Embarked : Port of Embarkation, C = Cherbourg, Q = Queenstown, S = Southampton

# Import Seaborn

```python
import seaborn as sns
```

# Q9: Draw barplot represents survived people based on sex

```python
sns.barplot(x="sex",y="survived",data=df)
plt.show()
```

# Q10: Draw catplot represents survived people based on embarked

```python
sns.catplot(x='embarked', hue='survived', kind='count', data=df)
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.title('Survived People based on Embarked')
plt.show()
```

```python
sns.catplot(x= 'survived', col= 'embarked', kind = 'count', data = df)
plt.show()
```

```python
df_embarked_s = df[df['embarked'] == 'S']
sns.catplot(x= 'survived', col= 'embarked', kind = 'count', data = df_embarked_s)
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.title('Survived People based on Embarked (Embarked = S)')
plt.show()
```

```python
df_embarked_c = df[df['embarked'] == 'C']
sns.catplot(x= 'survived', col= 'embarked', kind = 'count', data = df_embarked_c)
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.title('Survived People based on Embarked (Embarked = C)')
plt.show()
```

```python
df_embarked_q = df[df['embarked'] == 'Q']
sns.catplot(x= 'survived', col= 'embarked', kind = 'count', data = df_embarked_q)
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.title('Survived People based on Embarked (Embarked = Q)')
plt.show()
```

# Q11: Draw boxplot represents distribution of male and female based on age and pclass

```python
plt.figure(figsize=(15,8))
sns.boxplot(x = 'sex', y = 'age', hue = 'pclass', data = df)
plt.show()
```

# Q12: Draw heatmap represents correlations between sibsp, parch, age, fare, and survived columns

```python
columns = ['sibsp', 'parch', 'age', 'fare', 'survived']
subset_df = df[columns]
correlation_matrix = subset_df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

# Q13: Draw factorplot represents the relation between sibsp and survived columns

```python
sns.catplot(x='sibsp', y='survived', kind='point', data=df)
plt.xlabel('Number of Siblings/Spouses')
plt.ylabel('Survived')
plt.title('Survival Rate based on Number of Siblings/Spouses')
plt.show()
```

# Q14: Draw extra insights [Optional]

```python
age_data = df['age']
plt.figure(figsize=(8, 6))
sns.kdeplot(age_data, shade=True)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution (KDE)')
plt.show()
```
