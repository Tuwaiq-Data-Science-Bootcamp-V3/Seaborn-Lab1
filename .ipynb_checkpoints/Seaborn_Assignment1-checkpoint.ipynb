## Data Description

- Survival : 0 = No, 1 = Yes
- pclass(Ticket Class) : 1 = 1st, 2 = 2nd, 3 = 3rd
- Sex(Gender) : Male, Female
- Age : Age in years
- SibSp : Number of siblings/spouses abroad the titanic
- Parch : Number of parents/children abrod the titanic
- Ticket : Ticket Number
- Fare : Passenger fare
- Cabin : Cabin Number
- Embarked : Port of Embarkation, C = Cherbourg, Q = Queenstown, S = Southampton

# Q4: Add a new_column to a DataFrame that combines class and embark_town

```python
df['new_column']=df['class']+ " "+ df['embark_town']
df.head(5)
```

# Q5: Remove new_column from the DataFrame

```python
dataframe = df.drop('new_column', axis=1)
dataframe.head(5)
#another way 
#df.drop(columns=['new_column'])

#df.head(5)
```

# Q6: Filter DataFrame for rows of survived Males only 

```python
dataframe[(dataframe['sex'] =='male' ) &(dataframe['survived']==1)]
#another way 
#dataframe.query("sex == 'male' and survived==1")
```

# Q7: The total number of males who survived 

```python

total_number = dataframe[(dataframe['sex'] == 'male') & (dataframe['survived'] == 1)]
print('survived    ',total_number['survived'].count())
print('dtype:',total_number['survived'].dtypes)

```

# Import Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Q9: Draw barplot represents survived people based on sex

```python
plt.figure(figsize=(6,3))

sns.barplot(x='sex', y='survived',data=dataframe); # 
```

# Q10: Draw catplot represents survived people based on embarked

```python
sns.catplot(
    data=dataframe, x="survived", col="embarked",
    kind="count", aspect=1,
);
```

# Q11: Draw boxplot represents distribution of male and female based on age and pclass

```python
plt.figure(figsize=(11,6))
sns.boxplot(x='sex', y='age', hue='pclass', data=dataframe);
```

# Q12: Draw heatmap represents correlations between sibsp, parch, age, fare, and survived columns

```python
cols = ['sibsp', 'parch', 'age', 'fare', 'survived']
corr = dataframe[cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', square=True);

```

# Q13: Draw factorplot represents the relation between sibsp and survived columns

```python
#sns.factorplot(x='sibsp', y='survived', data=dataframe)# not working 
sns.catplot(x='sibsp', y='survived', data=dataframe, kind='bar');

```
